```python
import math
 
#zad1
 
def blad_wzgledny(rzecz, aproks):
    wynik = abs((rzecz - aproks) / rzecz)
    return wynik
   
def blad_bezwzgledny(rzecz, aproks):
    wynik = abs(rzecz - aproks)
    return wynik
 
#1
p = math.pi
p1 = 22/7
print("Blad wzgledny wynosi ", blad_wzgledny(p, p1))
print("Blad bezwzgledny wynosi ", blad_bezwzgledny(p, p1))
 
#2
p = math.exp(1)
p1 = 2.718
print("Blad wzgledny wynosi ", blad_wzgledny(p, p1))
print("Blad bezwzgledny wynosi ", blad_bezwzgledny(p, p1))
 
#3
p = 10 ** math.pi
p1 = 1397
print("Blad wzgledny wynosi ", blad_wzgledny(p, p1))
print("Blad bezwzgledny wynosi ", blad_bezwzgledny(p, p1))
 
#4
p = math.factorial(9)
p1 = math.sqrt(18 * math.pi) * ((9 / math.exp(1)) ** 9)
print("Blad wzgledny wynosi ", blad_wzgledny(p, p1))
print("Blad bezwzgledny wynosi ", blad_bezwzgledny(p, p1))
 
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
#zad2
def ciag(n):
    result = []
    for p in range(1, n+1):
        result.append((1.0/p) * (math.pow(10.0, p)*(1 + p * round(math.pi, 15) * math.pow(10.0, -p)) - math.pow(10.0, p)))
    return result
 
wektor = ciag(20)
print(wektor)
 
#2.1
#Wartosc G(p) nie zalezy od p, licząc analitycznie, po wymnożeniu nawiasów i skróceniu pewnych wyrażeń 
#zostaje liczba pi
 
#2.2
bezw = []
wzgl = []
for i in wektor:
    bezw.append(abs(i - round(math.pi, 15)))
    wzgl.append(abs((i - round(math.pi, 15)) / round(math.pi, 15)))
   
print("Blad bezwzgledny: ",bezw)
print("Blad wzgledny: ",wzgl)

x = np.linspace(0,20,20)
plt.semilogy(x, bezw)
plt.title('Błąd bezwzględny w zależności od n')
plt.xlabel('n')
plt.ylabel('błąd bezwzględny')
plt.legend('Wykres 1')
plt.show()

x = np.linspace(0,20,20)
plt.semilogy(x, wzgl)
plt.title('Błąd względny w zależności od n')
plt.xlabel('n')
plt.ylabel('błąd względny')
plt.legend('Wykres 2')
plt.show()

#2.3 wraz ze wzrostem n błąd rośnie, więc jest zależny od n
#2.4 błąd zaokrąglenia
```

```python
#zad3
def przyblizenie_e(n):
    result = 0
    for i in range(0,n+1):
        result += 1/math.factorial(i)
    return result

e = math.exp(1)

#3.1
e1 = przyblizenie_e(5)
print("błąd względny: ",blad_wzgledny(e,e1))
print("błąd bezwzględny: ",blad_bezwzgledny(e,e1))

#3.2
e2 = przyblizenie_e(10)
print("błąd względny: ",blad_wzgledny(e,e2))
print("błąd bezwzględny: ",blad_bezwzgledny(e,e2))
```
